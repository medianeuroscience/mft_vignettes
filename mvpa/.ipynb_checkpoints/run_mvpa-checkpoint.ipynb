# Multivoxel Pattern Analysis 

This notebook documents the analysis reported in section _Moral foundations elicit dissociable cortical activation patterns_.

The code follows the forced-choice tests documented in the [nltools](https://nltools.org/auto_examples/02_Analysis/plot_multivariate_classification.html#sphx-glr-auto-examples-02-analysis-plot-multivariate-classification-py) library.

```python
sub_list = ["sub-" + "{0:0>2}".format(x) for x in range(1,65)]
subject_id = np.concatenate([sub_list,sub_list])
```

## Whole-Brain

Train SVM and perform forced-choice test using voxels from the whole brain. 

Note that this takes quite some time. 

```python
loso_wb_svm_stats = {}
loso_wb_roc_stats = {}

for conditions in list(combinations(avg_beta_imgs.keys(), 2)):
    print("Starting SVM for: ", conditions)
    ## ---- Whole-Brain SVM ---- ###
    data = avg_beta_imgs[conditions[0]].append(avg_beta_imgs[conditions[1]])
    Y = pd.DataFrame(np.hstack([np.ones(64), np.zeros(64)]))
    data.Y = Y
    # Run LOSO SVM
    svm_stat = data.predict(algorithm='svm', cv_dict={'type': 'loso','subject_id':subject_id}, 
                            **{'kernel':"linear"}, plot=False, verbose=0)
    loso_wb_svm_stats[conditions] = svm_stat
    
    # Forced-choice test ROC
    roc_fc = Roc(input_values=svm_stat['dist_from_hyperplane_xval'], 
                 binary_outcome=svm_stat['Y'].astype(bool), forced_choice=subject_id)
    
    roc_fc.calculate()
    loso_wb_roc_stats[conditions] = roc_fc
    print("Completed Whole-Brain SVM")
```

```python
with open('output/loso_wb_svm.pkl', 'wb') as handle:
    pickle.dump(loso_wb_svm_stats, handle)
with open('output/loso_wb_roc.pkl', 'wb') as handle:
    pickle.dump(loso_wb_roc_stats, handle)
```

```python
# Parcelation
brain_mask = Brain_Data('http://neurovault.org/media/images/2099/Neurosynth%20Parcellation_0.nii.gz')
brain_mask_x = expand_mask(brain_mask)

# Moral Rois
moral_rois = {
            4:'V1',
            35:'Primary Auditory', 
            0:'amPFC',
            32:'vmPFC',
            2:'dmPFC',
            44:'dlPFC',
            22:'dACC',
            5:'TPJ/Ang. Gyrus',
            15:'TPJ/Par. Operculum',
            6:'PCC/Precuneus',
            19:'PCC/Superior LOC',
            49:'STS',
            16:'mInsula',
            18:'daInsula',
            37:'vaInsula'
}

moral_rois_mask = collapse_mask(brain_mask_x[list(moral_rois.keys())])
moral_rois_mask_x = expand_mask(moral_rois_mask)
moral_rois_mask.plot()
```

## ROIs

```python
loso_roi_svm_stats = {k:{} for k in moral_rois.values()}
loso_roi_roc_stats = {k:{} for k in moral_rois.values()}

for conditions in list(combinations(avg_beta_imgs.keys(), 2)):
    print("Starting SVM for: ", conditions)
    ## ---- ROI SVM ---- ###
    data = avg_beta_imgs[conditions[0]].append(avg_beta_imgs[conditions[1]])
    Y = pd.DataFrame(np.hstack([np.ones(64), np.zeros(64)]))
    data.Y = Y
    svm_stat_rois = data.predict_multi(algorithm='svm', method='rois',rois=moral_rois_mask_x, n_jobs=-1,
                                   cv_dict={'type': 'loso','subject_id':subject_id}, **{'kernel':"linear"})
        
    for ix, roi in enumerate(moral_rois.values()):
        loso_roi_svm_stats[roi][conditions] = svm_stat_rois[ix]
        # Forced-choice test ROC
        roc_fc = Roc(input_values=svm_stat_rois[ix]['dist_from_hyperplane_xval'], 
                 binary_outcome=svm_stat_rois[ix]['Y'].astype(bool), forced_choice=subject_id)
    
        roc_fc.calculate()
        loso_roi_roc_stats[roi][conditions] = roc_fc
    print("Completed ROI SVM")
```

In the following section, a multiclass decoder is developed using the same beta estimate maps as before, but only voxels from the Neurosynth mask. 

# Within-Subject Classification across Vignettes

Perform classification for a range of classifiers to determine which was yields highest accuracy.

```python
beta_maps = [x for x in glob.glob('betas/runwise_avg_smooth_zscored/sub-*') if x.split('_')[-4].split('/')[1] != 'sub-35']
beta_maps.sort()

# Split the beta maps into individual lists corresponding to the images (24) of each 
# subject 
lol = lambda lst, sz: [lst[i:i+sz] for i in range(0, len(lst), sz)]
all_maps = lol(beta_maps, 24)

# Loop over beta images per subject and perform decoding
def ws_decoder(sub_list):    
    # Get current sub_id
    sub_id = sub_list[0].split('_')[-4].split('/')[1]
    # Get condition and session (run) labels
    conditions_label = [x.split("_")[-3] for x in sub_list]
    session_label = [int(x.split("_")[-1].split('.')[0]) for x in sub_list]
    
    accs = pd.DataFrame(columns=['subject']+list(set(conditions_label))+['Average']+['Classifier'])
    decoder = Decoder(estimator='svc_l1', 
                      mask=neurosynth_moral.to_nifti(), 
                      standardize=False,cv=LeaveOneGroupOut(), scoring='roc_auc')
    # Fit the decoder
    decoder.fit(sub_list, conditions_label, groups=session_label)

    # Average Accuracy across ALL classes
    classification_accuracy = np.mean(list(decoder.cv_scores_.values()))
    # Averaged accuracy for each class
    mean_accs = pd.DataFrame(decoder.cv_scores_).mean()
    mean_accs = pd.DataFrame(mean_accs).T
    mean_accs['Average'] = float(classification_accuracy)
    mean_accs['subject'] = sub_id
 
    return mean_accs
```

```python
ws_accs = Parallel(n_jobs=20)(delayed(ws_decoder)(sub) for sub in all_maps)
ws_accs = pd.concat(ws_accs).reset_index(drop=True)
ws_accs_long = ws_accs.melt(id_vars=['subject'], var_name='Condition', value_name='ROC AUC')
```

```python
fig, ax = plt.subplots(figsize=(8,5))

label_order = ['Physical Care','Emotional Care', 'Fairness', 'Liberty','Loyalty','Authority','Sanctity', 'Social', 'Average']

sns.barplot(x='Condition', y='ROC AUC', data=ws_accs_long, order=label_order, palette='colorblind',
            errcolor='.1')

ax.set_xticklabels(label_order, rotation=20)
ax.set_ylabel('ROC AUC', weight='bold', fontsize=20)
ax.set_xlabel('')
sns.despine()
plt.ylim(0.5, 1)
ax.tick_params(axis='y', labelsize=15)
ax.tick_params(axis='x', labelsize=12)
plt.title('Within-Subject Classification', weight='bold', fontsize=20)
plt.savefig('plots/ws_accs.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Decoding and Cross-Decoding of Moral Wrongness Ratings

```python
cond_labels = ['Physical Care', 'Emotional Care', 'Fairness', 'Liberty', 'Loyalty','Authority','Sanctity','Social']
```

```python
svc_ovo = OneVsOneClassifier(
    Pipeline(
        [
            ("svc", LinearSVC(penalty="l1", dual=False)),
        ]
    )
)
```

```python
cv = LeaveOneGroupOut()
```

## Neurosynth Moral

```python
beta_maps = [x for x in glob.glob('/srv/lab/fmri/mft/fhopp_diss/analysis/vignettes/mvpa/betas/new_condition_ratings_zscored/sub-*')]
beta_maps.sort()

y = pd.Series([x.split('/')[-1].split('_')[1] for x in beta_maps])
subs = pd.Series([int(x.split('/')[-1].split('_')[0].split('-')[1]) for x in beta_maps])
```

```python
X = Brain_Data(beta_maps).apply_mask(neurosynth_moral).data
```

### Cross-Decoding of Moral Wrongness

```python
def cv_cross_decode(train, test, X, y, cond):    
    # Get index for all but held-out condition map 
    hold_out_map_ixs = pd.Series(y)[pd.Series(y) != cond].index
    # Select training for all samples except held out foundation 
    train = [x for x in train if x in hold_out_map_ixs]
    # Select test for only held out foundation
    test = [x for x in test if x not in hold_out_map_ixs]
    
    X_train, y_train = X[train], y.iloc[train]
    
    svc_ovo.fit(X_train, y_train)
    
    X_test, y_test = X[test], y.iloc[test]
    y_pred = svc_ovo.predict(X_test)
    
    cm_norm = confusion_matrix(y_test, y_pred, labels=cond_labels, normalize='true')
    cm_raw = confusion_matrix(y_test, y_pred, labels=cond_labels)
    
    return cm_norm, cm_raw
```

```python
cm_normed = []; cm_counts = []
for cond in cond_labels:
    # Raw and normalized confusion matrices 
    bsc_pred_scores = Parallel(n_jobs=-1)(delayed(cv_cross_decode)(train, test, X, y, cond) for (train, test) in cv.split(X, y, subs))
    normed_cms = [x[0] for x in bsc_pred_scores]
    count_cms = np.sum([x[1] for x in bsc_pred_scores], axis=0)
    cm_counts.append(count_cms)
    # Average normalized confusion matrices across subjects
    mean_bsc_acc = np.mean(normed_cms, axis=0)
    cm_normed.append(mean_bsc_acc)
```

```python
sum_bsc_cm = pd.DataFrame(sum(cm_counts))
mean_bsc_cm = sum(cm_normed)
```

```python
p_vals_cm = sum_bsc_cm.copy().astype(str)
for col in p_vals_cm.columns:
    p_vals_cm[col].values[:] = ''
    
for row_ix, row in sum_bsc_cm.iterrows():
    for col_ix, count in row.iteritems():
        # binomal test 
        p_val = binom_test(x=count, n=64, p=1/7, alternative='greater')
        if p_val < 0.001:
            p_vals_cm.at[row_ix, col_ix] = '*'
```

```python
sns.heatmap(mean_bsc_cm, cmap='turbo', square=True,annot=p_vals_cm, fmt="s",annot_kws={'size': 20},
           xticklabels=cond_labels, yticklabels=cond_labels)
plt.title('Neurosynth Moral \n Cross-Decoding of Moral Wrongness', fontsize=18)
plt.xticks(fontsize=15)
plt.ylabel('Held-Out Condition', fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Predicted Condition', fontsize=15)
plt.savefig('plots/ns_cross_decoding.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Decoding of Moral Wrongness

```python
def cv_decode(train, test, X, y):        
    X_train, y_train = X[train], y.iloc[train]
    
    svc_ovo.fit(X_train, y_train)
    
    X_test, y_test = X[test], y.iloc[test]
    y_pred = svc_ovo.predict(X_test)
    
    cm_norm = confusion_matrix(y_test, y_pred, labels=cond_labels, normalize='true')
    cm_raw = confusion_matrix(y_test, y_pred, labels=cond_labels)
    
    return cm_norm, cm_raw
```

```python
# Raw and normalized confusion matrices 
bsc_pred_scores = Parallel(n_jobs=-1)(delayed(cv_decode)(train, test, X, y) for (train, test) in cv.split(X, y, subs))
normed_cms = [x[0] for x in bsc_pred_scores]
# Average normalized confusion matrices across subjects
mean_bsc_acc = np.mean(normed_cms, axis=0)
```

```python
sum_bsc_cm = pd.DataFrame(np.sum([x[1] for x in bsc_pred_scores], axis=0))
p_vals_cm = sum_bsc_cm.copy().astype(str)
for col in p_vals_cm.columns:
    p_vals_cm[col].values[:] = ''
    
for row_ix, row in sum_bsc_cm.iterrows():
    for col_ix, count in row.iteritems():
        # binomal test 
        p_val = binom_test(x=count, n=64, p=1/8, alternative='greater')
        if p_val < 0.001:
            p_vals_cm.at[row_ix, col_ix] = '*'
```

```python
sns.heatmap(mean_bsc_acc, cmap='turbo', square=True, annot=p_vals_cm, fmt="s",annot_kws={'size': 20},
           xticklabels=cond_labels, yticklabels=cond_labels,)
plt.title('Neurosynth Moral \n Decoding of Moral Wrongness', fontsize=18)
plt.xticks(fontsize=15)
plt.ylabel('Actual Condition', fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Predicted Condition', fontsize=15)
plt.savefig('plots/ns_decoding.png', dpi=300, bbox_inches='tight')
plt.show()
```
