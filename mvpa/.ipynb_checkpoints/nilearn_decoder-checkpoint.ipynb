# Vignettes Decoder

In this notebook, a decoder is developed that can learn to distinguish the neural patterns associated with each moral foundation as well as social norms. I will use statistical maps from a first level GLM and limit the features (voxels) to a Neurosynth mask that loads on the term "moral". 

```python
from nilearn.decoding import Decoder
from sklearn.model_selection import LeaveOneGroupOut
from nltools import Brain_Data

from nilearn.plotting import plot_glass_brain

import glob
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import ttest_1samp, ttest_ind
from nltools.data import Brain_Data
from nltools.mask import expand_mask, collapse_mask

from joblib import Parallel, delayed
import pickle
```

```python
moral_mask = Brain_Data("../glm/masks/moral_association-test_z_FDR_0.01.nii.gz").threshold(1, binarize=True)
```

```python
moral_mask.plot()
```

# Within-Subject Classification across Vignettes

Perform classification for a range of classifiers to determine which was yields highest accuracy.

```python
beta_maps = [x for x in glob.glob('../mvpa/betas/runwise_betas/sub-*') if x.split('_')[-4].split('/')[1] != 'sub-35']
beta_maps.sort()

# Split the beta maps into individual lists corresponding to the images (24) of each 
# subject 
lol = lambda lst, sz: [lst[i:i+sz] for i in range(0, len(lst), sz)]
all_maps = lol(beta_maps, 24)

# Loop over beta images per subject and perform decoding
def ws_decoder(sub_list):    
    # Get current sub_id
    sub_id = sub_list[0].split('_')[-4].split('/')[1]
    # Get condition and session (run) labels
    conditions_label = [x.split("_")[-3] for x in sub_list]
    session_label = [int(x.split("_")[-1].split('.')[0]) for x in sub_list]
    # Init. Decoder for different classifiers
    classifiers = ['svc_l2', 'svc_l1', 'logistic_l1',
               'logistic_l2', 'ridge_classifier']
    
    accs = pd.DataFrame(columns=['subject']+list(set(conditions_label))+['Average']+['Classifier'])
    for i, classifier_name in enumerate(sorted(classifiers)):
        decoder = Decoder(estimator=classifier_name, 
                          mask=moral_mask.to_nifti(), 
                          standardize=False,cv=LeaveOneGroupOut())
        # Fit the decoder
        decoder.fit(sub_list, conditions_label, groups=session_label)

        # Average Accuracy across ALL classes
        classification_accuracy = np.mean(list(decoder.cv_scores_.values()))
        # Averaged accuracy for each class
        mean_accs = pd.DataFrame(decoder.cv_scores_).mean()
        # Store results
        accs.at[i, 'subject'] = sub_id
        accs.at[i, 'Average'] = float(classification_accuracy)
        accs.at[i, 'Classifier'] = classifier_name
        

        for label, acc in mean_accs.iteritems():
            accs.at[i, label] = float(acc)
            
    return accs
```

```python
ws_accs = Parallel(n_jobs=8)(delayed(ws_decoder)(sub) for sub in all_maps)
ws_accs = pd.concat(ws_accs).reset_index(drop=True)
```

```python
ws_accs_long = ws_accs.melt(id_vars=['subject','Classifier'], var_name='Condition', value_name='Accuracy')
```

### Plot Outcomes of Different Classifiers (not reported in the paper)

```python
fig, ax = plt.subplots(figsize=(12,5))

label_order = ['Physical Care','Emotional Care', 'Fairness', 'Liberty','Loyalty','Authority','Sanctity', 'Social', 'Average']

sns.barplot(x='Condition', y='Accuracy', hue='Classifier', data=ws_accs_long, order=label_order, alpha=0.6, errcolor='.1')

ax.set_xticklabels(label_order, rotation=20)
plt.gca().axhline(1/8, color='black', linestyle='--')
ax.set_ylabel('Classification Accuracy (AUC Score)', weight='bold')
ax.set_xlabel('')
sns.despine()
plt.title('Within-Subject Classification \n Across 8 MFV Conditions', weight='bold')
plt.savefig('plots/ws_accs_classifiers.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Plot Winner (SVC_L1)

```python
svc_l1 = ws_accs[ws_accs['Classifier']=='svc_l1']
del svc_l1['Classifier']
svc_l1_accs_long = svc_l1.melt(id_vars=['subject'], var_name='Condition', value_name='Accuracy')
```

```python
fig, ax = plt.subplots(figsize=(8,5))

label_order = ['Physical Care','Emotional Care', 'Fairness', 'Liberty','Loyalty','Authority','Sanctity', 'Social', 'Average']

sns.barplot(x='Condition', y='Accuracy', data=svc_l1_accs_long, order=label_order, alpha=0.6, palette='dark', errcolor='.1')
sns.stripplot(x='Condition', y='Accuracy', data=svc_l1_accs_long, order=label_order, alpha=0.3, color='black', dodge=True, size=3)

ax.set_xticklabels(label_order, rotation=20)
plt.gca().axhline(1/8, color='black', linestyle='--')
ax.set_ylabel('Classification Accuracy (AUC Score)', weight='bold')
ax.set_xlabel('')
sns.despine()
plt.title('Within-Subject Classification \n Across 8 MFV Conditions', weight='bold')
plt.savefig('plots/ws_accs.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Perform One-Sample T-tests again chance accs
for label in ['Physical Care','Emotional Care', 'Liberty','Fairness','Loyalty','Authority','Sanctity', 'Social','Average']:
    print(label, ttest_1samp(svc_l1[label], popmean=1/8))
```

# Between-Subject Classification across Vignettes

```python
beta_maps = [x for x in glob.glob('../mvpa/betas/averaged_betas/sub-*')]
beta_maps.sort()

# Split the beta maps into individual lists corresponding to the images (8) of each 
# subject 
lol = lambda lst, sz: [lst[i:i+sz] for i in range(0, len(lst), sz)]
all_maps = lol(beta_maps, 8)

conditions_label = [x.split("_")[-2] for x in beta_maps]
sub_label = [int(x.split("/")[-1].split('_')[0].split('-')[1]) for x in beta_maps]
```

```python
decoder = Decoder(estimator='svc_l1', 
                      mask=moral_mask.to_nifti(), 
                      standardize=False,cv=LeaveOneGroupOut(), n_jobs=8, verbose=1)
decoder.fit(beta_maps, conditions_label, groups=sub_label)
```

```python
with open('bsc_decoder.pkl', 'wb') as outfile:
    pickle.dump(decoder, outfile)
```

```python
bs_accs = pd.read_pickle('bsc_decoder.pkl')
```

```python
bs_accs = pd.DataFrame(bs_accs.cv_scores_)
bs_accs['Average'] = bs_accs.mean(axis=1)
bs_accs = bs_accs.reset_index()
```

```python
bs_accs_long = bs_accs.melt(id_vars=['index'], var_name='Condition', value_name='Accuracy')
```
