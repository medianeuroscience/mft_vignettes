# Visualization of Statistical Maps on Cortical Surface via Surfplot

```python
from dipy.viz import window
from xvfbwrapper import Xvfb

from neuromaps import transforms
from surfplot import Plot
from neuromaps.datasets import fetch_fslr
```

```python
vdisplay = Xvfb()
vdisplay.start()
```

```python
second_lv_maps = '/srv/lab/fmri/mft/fhopp_diss/analysis/vignettes/glm/spm/second_lv/datasink/2ndLevel/'
```

```python
contrast_mapper = {
    'con_0018':'Physical Care > Social',
    'con_0019':'Emotional Care > Social',
    'con_0020':'Fairness > Social',
    'con_0021':'Liberty > Social',
    'con_0022':'Loyalty > Social',
    'con_0023':'Authority > Social',
    'con_0024':'Sanctity > Social',
    'con_0025':'Binding > Individualizing',
    'con_0026':'Moral > Social'
}
```

```python
surfaces = fetch_fslr()
lh, rh = surfaces['inflated']
```

```python
# Kept for legacy. 

for contrast,name in contrast_mapper.items():

    stat_img = second_lv_maps+f'{contrast}/spmT_0001_thr.nii'
    fslr = transforms.mni152_to_fslr(stat_img, '32k')
    lh_data, rh_data = fslr
    
    p = Plot(surf_lh=lh, surf_rh=rh)
#     fig = p.build()

    sulc_lh, sulc_rh = surfaces['sulc']
    p.add_layer({'left': sulc_lh, 'right': sulc_rh}, cmap='binary_r', cbar=False)
    
    p.add_layer({'left': lh_data, 'right': rh_data}, cmap='YlOrRd_r')
    fig = p.build(cbar_kws={'fontsize': 15,'decimals': 1,'aspect': 8,'draw_border': False,
                           'pad':0})
    fig.suptitle(name, fontsize=15, fontweight='bold')
#     fig.savefig(f'plots/second_lv_{name}.png',dpi=300,bbox_inches='tight', pad_inches=0)
```

## Conjunction Map

```python
fslr = transforms.mni152_to_fslr('../second_lv/datasink/2ndLevel/conjunction.nii', '32k')
lh_data, rh_data = fslr

p = Plot(surf_lh=lh, surf_rh=rh)

sulc_lh, sulc_rh = surfaces['sulc']
p.add_layer({'left': sulc_lh, 'right': sulc_rh}, cmap='binary_r', cbar=False)

p.add_layer({'left': lh_data, 'right': rh_data}, cmap='YlOrRd_r', cbar=False)
fig = p.build()
fig.suptitle("Conjunction", fontsize=15, fontweight='bold')
# fig.savefig(f'plots/second_lv_conjunction.png',dpi=300,bbox_inches='tight', pad_inches=0)
```

## Neurosynth Maps

```python
moral_map = '/srv/lab/fmri/mft/fhopp_diss/analysis/vignettes/glm/spm/notebooks/neurosynth_maps/moral_uniformity-test_z_FDR_0.01.nii.gz'
```

```python
surfaces = fetch_fslr()
lh, rh = surfaces['inflated']
```

```python
fslr = transforms.mni152_to_fslr(moral_map, '32k')
lh_data, rh_data = fslr
p = Plot(surf_lh=lh, surf_rh=rh)
sulc_lh, sulc_rh = surfaces['sulc']
p.add_layer({'left': sulc_lh, 'right': sulc_rh}, cmap='binary_r', cbar=False)
p.add_layer({'left': lh_data, 'right': rh_data}, cmap='YlOrBr_r',  cbar_label='Moral')

kws = {'location': 'right', 'label_direction': 0, 'decimals': 1,
       'fontsize': 12, 'n_ticks': 4, 'aspect': 8,
       'draw_border': False}

fig = p.build(cbar_kws=kws)
# fig.savefig(f'plots/neurosynth_maps.png',dpi=300,bbox_inches='tight', pad_inches=0)
```
